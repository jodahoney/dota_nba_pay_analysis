```python
# finish webscraping ages of death, takes about 10 mins to run
import wikipedia
import numpy

wikipedia.set_rate_limiting(True) # limit queries to default once every 50 ms
unique_names = athlete_events_df.Name.unique()
temp_df = athlete_events_df.copy()

for name_raw in unique_names:
  # clean up the query string for better suggestion results
  name = name_raw.replace('\"',' ') # remove double quotes
  name = name.replace('(',' ') # remove parantheses
  name = name.replace(')',' ') # remove parantheses
  name = name.replace('-',' ') # remove hyphens
  
  # try to find athlete's wiki page
  try:
    wikipage = wikipedia.page(title=name, pageid = None, auto_suggest=True)
  except:
    # if not found, overwrite athlete's name with NaN in new dataframe
    temp_df.loc[temp_df.Name == name_raw, 'Name'] = numpy.nan 
    continue
    
  # if athlete's wiki page is in, e.g., '1970 deaths' category
  if len(wikipage.categories)>1 and 'deaths' in wikipage.categories[1]: 
    death_year = int(wikipage.categories[1][0:4])
  else:
    death_year = None
  if death_year:
    print(name,'year of death found...')
    # year of death found
    temp_df.loc[temp_df.Name == name_raw, 'Name'] = death_year
  else:
    # remove athlete's name in new dataframe
    temp_df.loc[temp_df.Name == name_raw, 'Name'] = numpy.nan 

# rename Name column
ages_events_df = temp_df.rename(columns = {'Name': 'death_year'}) 
```

Cleaning up every nba_salary dataset

Changing index to ranking

Merging all nba_salary datasets to get a seperate overall earnings dataset of NBA players

```python
#Cleaning up every nba_salary dataset
#Changing index to ranking
#Removing position played from name as unnecessary data
#Add unit to Salary column
#Remove Team played on since unnecessary data

for df_name in outputdict:
  if 'nba_salary' in df_name: #Checks if current dataframe is an 'nba_salary' dataframe
    temp_df = outputdict[df_name].copy() #Make copy of dataframe to clean
    temp_df  = temp_df.set_index('RK')   #Set index of dataframe to 'RK' which is the ranking based on salary of players
    temp_df['NAME'] = temp_df['NAME'].apply(lambda x: x.split(',', 1)[0]) #Remove the position the player plays in and leaves just their names as the position is not relevant data and makes this column more uniform across dataframes as players may change posiitons in a different year
    temp_df = temp_df.rename({'SALARY': 'SALARY (US$)'}, axis='columns') #Add currency unit to salary column name
    temp_df = temp_df.drop(['TEAM'], axis=1) #Drops team column as we do not need this data and it is just extra data taking up space in our process
    outputdict[df_name] = temp_df #Reassign the current dataframe in dictionary of dataframes to cleaned one

#Merging all nba_salary datasets to get an overall earnings dataset of NBA players
merge_df = pd.DataFrame() #Create empty dataframe to fill
for df_name in outputdict:
  if 'nba_salary' in df_name: #Checks if current dataframe is an 'nba_salary' dataframe
    curr_df = outputdict[df_name].copy() #Make a copy of dataframe to work on
    if merge_df.empty:
      merge_df = curr_df #Assign empty dataframe to current dataframe to set as base dataframe
    else:
      merge_df = merge_df.set_index('NAME').add(curr_df.set_index('NAME'), fill_value=0).reset_index() #Set merge_df and curr_df index to player name then add curr_df to merge_df with condition that if player name isn't found in curr_df to merge_df to set unfound value to 0
merge_df = merge_df.sort_values(by=['SALARY (US$)'], ascending=False).set_index('NAME') #Sort merge_df by salary after merging with all nba_salary dataframes
outputdict['nba_overall_earnings'] = merge_df #Add merge_df to dictionary of dataframes with name 'nba_overall_earnings' so we can compare to our dataframe of overall earnings for Dota2 players

print(outputdict['nba_overall_earnings'])
```

```python
#Obtain names from URL used to scrape data from and removed unnecessary columns while renaming the remaining columns to the appropriate name
#Grouped earnings by year and created dictionary of dataframes that is indexed by years
#Added this to dictionary of dataframes as well as an overall earnings of every player dataframe
#Added units to monetary values column and converted the strings to floats so quantitative operations may be done on them
#Cut off 2019 earnings because the year is not over so cannot get their total earnings of 2019, this is also why 2019 salary of NBA players not been included
#Note : The first part of 'Name' is each respective players username

temp = outputdict['dota_2_player_earnings'].copy() #Make copy of dataframe to work on
temp['URL'] = temp['URL'].apply(lambda x: x.replace('https://www.esportsearnings.com/players/','')).apply(lambda x: " ".join(x.split('-')[1:])) #Get player names from the URLs obtained from data scraping
temp = temp.rename({'URL':'Name', 'Data.Column1': 'Year', 'Data.Column2' : 'Total Results (US$)'}, axis=1) #Rename columns to more appropriate names by inspection
cols = ['Name','Year','Total Results (US$)']
drops = [c for c in temp.columns if c not in cols] #Get list of unnecessary columns
i = temp[temp['Total Results (US$)'] == 'Total Results'].index 
temp = temp.drop(drops,axis=1) #Drop unnecessary columns
temp = temp.drop(i)
temp['Total Results (US$)'] = temp['Total Results (US$)'].apply(lambda x: float(x.replace("$","").replace(",",""))) #Remove ',' and '$' from monetary values and convert to float for ease of calculation use
df_groupby_years = {k: v for k, v in temp.groupby('Year')} #Create new dictionary of dataframe indexed by year of player earnings in that year
del df_groupby_years['2019'] #Since 2019 has not concluded, unable to get true earnings of players in 2019 so all earnings in 2019 are ommitted and  2019 salary of NBA players are not included
for k in df_groupby_years:
  df_groupby_years[k]= df_groupby_years[k].set_index("Name").drop(['Year'], axis=1) #Remove year column as year is being used as index already
  print(k) 
  print(df_groupby_years[k])
outputdict['dota_2_player_earnings_by_year'] = df_groupby_years #Add dictionary to dictionary of dataframes as dota_2_player_earnings_by_year will be used to compare to NBA player yearly earnings
temp = temp.drop(['Year'], axis=1) #Since Year column not necessary anymore (it was only there to be used for grouping to make dota_2_player_earnings_by_year) it is dropped
temp = temp.groupby(['Name']).sum().sort_values(by=['Total Results (US$)'],ascending=False) #dota_2_player_earnings is grouped by player name then earnings are summed then the dataframe is sorted by salary which will be used to compare to NBA player earnings overall
outputdict['dota_2_player_earnings'] = temp #Reassign dota_2_player_earnings to cleaned dataframe
print(outputdict['dota_2_player_earnings'])
```
